# Flipkart Web Scrapping
Here we are going to scrap Flipkart websites into Laptop.


Here what we will do this same process will be repeated to scrape other sites.



```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
r = requests.get(url)
print(r)
```

Response 200 means everythings went well 

```python
soup = BeautifulSoup(r.text, "lxml")
print(soup)
```

Here HTML file is scraped.

Now we have to scrape the next page so we have to find the HTML format to see it we have to inspect the Flipkart websites.

```python
np = soup.find("a", class_ = "_1LKTO3").get("href") #np means next page
print(np)
```

```python
cnp = "https://www.flipkart.com"+np #complete next page
print(cnp)
```

Now we have to scrape the url of next page till end so we have to write upper code in loop.

```python
while True:
    np = soup.find("a", class_ = "_1LKTO3").get("href") #np means next page
    cnp = "https://www.flipkart.com"+np
    print(cnp)

    url = cnp
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "lxml")
```

Here we get only two page one by one. First it print page 1 and then it prints page 2.
If we see all pages links we see that all pages of links are same only last string are changing from 1 to 2 to 3 and so on.

```python
for i in range(2, 10):
    url = "https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page="+str(i) #Here we have added i as str because here all page are same only string at laast are changing as 1,2,3 and so on. 
    r = requests.get(url)
    
    soup = BeautifulSoup(r.text, "lxml")
    
    np = soup.find("a", class_ = "_1LKTO3").get("href") #np means next page
    
    cnp = "https://www.flipkart.com"+np
    print(cnp)
```

Here we have got the 8 page links and then we have to get all page links and to scrape the data from it.

Now we are going to get the data of the page. we will remove loop to see data of only first page and then we will get data of others page.

```python
Product_name = []
Prices = []
Description = []


for i in range(2, 56):
    url = "https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page="+str(i)
    r = requests.get(url)

    soup = BeautifulSoup(r.text, "lxml")
    box = soup.find("div", class_ = "_1YokD2 _3Mn1Gg")

    names = box.find_all("div",class_ = "_4rR01T")

    for i in names:
        name = i.text
        Product_name.append(name)
        
    prices = box.find_all("div",class_= "_30jeq3 _1_WHN1")

    for i in prices:
        price = i.text
        Prices.append(price)
        
    desc = box.find_all("ul",class_= "_1xgFaf")

    for i in desc:
        des = i.text
        Description.append(des)
```

Here we are not scrapping review because some product don't have any review but this is not good practice to do. We will solve this problem in Future.

```python
df = pd.DataFrame({"Product_name":Product_name,"Prices":Prices,"Description":Description})
```
